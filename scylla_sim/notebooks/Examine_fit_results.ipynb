```python
import matplotlib.pyplot as plt
from astropy.table import QTable
```

```python
sim = "../beast_sim/scylla_sim_low_Av1.fits"
simtab = QTable.read(sim)

sim = "../beast_sim/scylla_sim_low_Av4.fits"
simtab3 = QTable.read(sim)
```

```python
print(simtab.colnames)
```

```python
plt.plot(simtab["Av"], simtab["Rv"], "ko")
```

```python
plt.plot(simtab3["Av"], simtab3["Rv"], "ko")
```

```python
plt.figure(figsize=(20, 10))
plt.plot(simtab["F475W_VEGA"] - simtab["F814W_VEGA"], simtab["F814W_VEGA"], "bo", label="Sim w/ lognorm A(V)=1")
plt.plot(simtab3["F475W_VEGA"] - simtab3["F814W_VEGA"], simtab3["F814W_VEGA"], "ro", label="Sim w/ lognorm A(V)=4")
plt.xlim(0.0, 6.0)
plt.ylim(28, 20)
plt.xlabel("F475W - F814W")
plt.ylabel("F814W")
plt.legend()
```
